```python
from sklearn.preprocessing import LabelEncoder
```

```python
le1 = LabelEncoder().fit(X_combined_data_edited['source'])
X_combined_data_edited['source_new'] = le1.transform(X_combined_data_edited['source'])
X_combined_data_edited = X_combined_data_edited.drop('source', axis=1)
```

```python
le1 = LabelEncoder().fit(X_combined_data_edited['financial_institution'])
X_combined_data_edited['financial_institution_new'] = le1.transform(X_combined_data_edited['financial_institution'])
X_combined_data_edited = X_combined_data_edited.drop('financial_institution', axis=1)
```

```python
le1 = LabelEncoder().fit(X_combined_data_edited['origination_date'])
X_combined_data_edited['origination_date_new'] = le1.transform(X_combined_data_edited['origination_date'])
X_combined_data_edited = X_combined_data_edited.drop('origination_date', axis=1)
```

```python
le1 = LabelEncoder().fit(X_combined_data_edited['first_payment_date'])
X_combined_data_edited['first_payment_date_new'] = le1.transform(X_combined_data_edited['first_payment_date'])
X_combined_data_edited = X_combined_data_edited.drop('first_payment_date', axis=1)
```

```python
le1 = LabelEncoder().fit(X_combined_data_edited['loan_purpose'])
X_combined_data_edited['loan_purpose_new'] = le1.transform(X_combined_data_edited['loan_purpose'])
X_combined_data_edited = X_combined_data_edited.drop('loan_purpose', axis=1)
```

# XGBoost

```python
from xgboost import XGBClassifier
```

```python
xgbclassifier = XGBClassifier(booster='dart', max_delta_step=5, reg_alpha=20, reg_lambda=100, scale_pos_weight=5,
              learning_rate=0.01, n_estimators=150, subsample=0.5)
```

```python
xgbclassifier.fit(train_edited, train_data['m13'])
```

```python
results = []
xbg_predict_proba = xgbclassifier.predict_proba(test_edited)
for i in xbg_predict_proba:
    if(i[0]>0.66):
        results.append(0)
    else:
        results.append(1)
results_df = pd.DataFrame(results, index=test_data['loan_id'], columns=['m13'])
```

```python
print(len(results_df[results_df['m13'] == 0]))
print(len(results_df[results_df['m13'] == 1]))
```
